# Ebay Germany Car Sale Data Exploration

This projects guide you through the cleaning and analysis of the [ebay car sale dataset.](https://www.kaggle.com/orgesleka/used-cars-database) We will use python based data analysis library: Pandas, Numpy, and seaborn to perform the exploratory data analysis of this dataset. 

```python
# Importing the required library
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Lets load the dataset and have a glimpse on it.

```python
data_file = "autos.csv"
# Since pandas default accepted encoding of csv is UTF-8, \
# we need to specify the encoding if it differs.
autos_data = pd.read_csv(data_file,encoding="Latin-1") 
print(autos_data.info())
autos_data
```

```python
autos_data.columns
```

```python
autos_data.rename(columns = {'offerType':'offer_type','price':'price_$','abtest':'ab_test','vehicletype':'vehicle_type', \
                             'yearOfRegistration':'year_of_registration','gearbox':'gear_box','powerPS':'power_ps', \
                             'odometer':'odometer_km','monthOfRegistration':'month_of_registration','fuelType':'fuel_type', \
                             'notRepairedDamage':'not_repaired_damage','dateCreated':'ad_created', \
                             'nrOfPictures':'nr_of_pictures','postalCode':'postal_code','lastSeen':'last_seen'},inplace=True)
autos_data.head(3)
```

```python
autos_data.describe(include='all')
```

From the summary above, we can see that offer_type and seller columns same data repetition and hence does not help in analysing the dataset.  So, we can delete this two columns. In addition,we noticed many non-text datatype stored as text and hence need data-type conversion to its correct format.

```python
autos_data.drop(["seller", "offer_type"], axis=1, inplace=True)
autos_data.shape
```

we will now update the price and odometer column to make the data numeric and analysable.

```python
autos_data['price_$'] = autos_data['price_$'].str.replace('$','')
autos_data['price_$'] = autos_data['price_$'].str.replace(',','')
autos_data['price_$'] = autos_data['price_$'].astype(float)
autos_data['price_$'].dtype
```

```python
autos_data['odometer_km'] = autos_data['odometer_km'].str.strip('km')
autos_data['odometer_km'] = autos_data['odometer_km'].str.replace(',','')
autos_data['odometer_km'] = autos_data['odometer_km'].astype(int)
autos_data['odometer_km'].dtype
```

Upon close observation, we can see the price range is very high (0.0 - 99999999.0).  Since, the  few max price are  unique and quite high compared to other prices.The prices of car in million available in ebay website is quite rare and hence we can consider the price above half million  as an outlier and thus remove them. Also the least price (0.0) is removed as an outlier.

```python
 autos_data= autos_data.loc[(autos_data['price_$'] > 0.0) & (autos_data['price_$'] < 500000 )] 
                                                          
```

```python
print(max(autos_data['price_$']))
print(min(autos_data['price_$']))
```

Since , we observed some incorrect values in 'year_of_registration' column, we will update those incorrect values. We will check the data and consider them outlier if they are before 1900 and after the current date. If these valuers are less than 5% (threshold value, we will considering removing the entire rows containing such value rather than performing other data imputing operations.

```python
1 - ((autos_data['year_of_registration'].between(1900,2016)).sum()/autos_data.shape[0])
```

There is only about 3.9% of data that seems to be outlier, which is less than our threshold value. Hence we can remove the rows that contains these values.

```python
autos_data.shape
```

```python
autos_data = autos_data[autos_data['year_of_registration'].between(1900,2016)]
autos_data.shape
```

Now, we will explore the various brands of cars.We find out there are fourty different brands of cars in our dataset. Lets check the dataset.

```python
brands_cars = autos_data['brand'].value_counts(normalize=True)
popular_brands = brands_cars[brands_cars > 0.10].index
popular_brands
```

There are only three brands that have more than 10% cars of it in this rbay car sales data. All of them are German based, of which, Volswagen is the frontrunner with more than a-fifth of the ebay car sales market. 

```python
cars_collection = {}

for brand in autos_data['brand']:
    if brand not in  cars_collection:
        cars_collection[brand] = 1
    else:
        cars_collection[brand] += 1
mean_value_cars_brands = pd.Series ([cars_collection[k] for k in cars_collection]).mean()
```

```python
popular_brands_price = {}

for  brand in popular_brands:
    brand_ = autos_data[autos_data['brand'] == brand]
    price = brand_['price_$'].mean()
    popular_brands_price[brand] = int(price)
popular_brands_price
```

```python
mean_prices_brand = autos_data.groupby('brand')['price_$'].mean()
mean_prices_brand = mean_prices_brand.astype(int)
```

We aggregate our popular car brands with their price value and get the mean price of easch brands. From the aggregation, we get that Opel cars seems to be quite cheaply available  as its mean price is less than 3000 whereas BMW is quite high and data reveals the brand value upon the mean price.  

```python
df = pd.DataFrame(mean_prices_brand)
df.head()
```

```python
del df.index.name
df.head()
```

```python
brands_mean_price = {}
brands_mean_mileage = {}

for brand in brands_cars.index:
    brand_ = autos_data[autos_data['brand'] == brand]
    price = brand_['price_$'].mean()
    mileage = brand_['odometer_km'].mean()
    brands_mean_price[brand] = int(price)
    brands_mean_mileage[brand] = int(mileage)    
```

```python
brands_mean_price = pd.Series(brands_mean_price)
brands_df = pd.DataFrame(brands_mean_price, columns=['mean_price'])
brands_mean_mileage = pd.Series(brands_mean_mileage)
brands_df['mean_mileage'] = brands_mean_mileage
brands_df.head()
```

```python
print(brands_df.corr())
print(brands_df.describe())
```

The correlation between the price tends to be negative  and very low.The average price of the various brands ranges from 1049 to  to 45643 Euros whereas the overall mean price tends be about 6400 contrary to median value(~4100e). This shows the skewness in price distribution.  The mileage distribution seems to be more normal compared to price.
