# Visualizing the College Majors based Earnings

In this project, we will analyse the job outcomes of the graduate from differnt majors in between 2010 and 2012. The raw data was collected by [American Community Survey](https://www.census.gov/programs-surveys/acs/),which was further cleaned and provided by [FiveThirtyEight](https://github.com/fivethirtyeight/data/tree/master/college-majors). 

We will mainly focus on data visualization and try to answer various questions bsased on the plots we create. let start the project by import all the required python libraries.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
file = 'recent-grads.csv'
recent_grads = pd.read_csv(file)
```

The above code verifies that missing values have been dropped. Now, lets visualize our data using the pandas plot method to understand the data further. 

```python
recent_grads.columns
```

```python
recent_grads.plot(x='Sample_size', y='Median',kind='scatter',title='Sample Size vs Median')
```

The plot shows that there is no clear correlation between the sample size(majors populasrity) and median  income.

```python
recent_grads.plot(x='Sample_size',y='Unemployment_rate',kind='scatter',title='Sample Size vs Unemployment Rate')
```

```python
recent_grads.plot(x='Full_time',y='Median',kind='scatter',title='Full time vs Median')
```

There is almost no correlation between full time employees and their median salary.

We can combine the scatter plot and histogram togeher and plot the correlation as well as individual features in the same plot using the panda's scatter_matrix plot method. First we will see the correlation between "sample size" and "Median" as well as their individual histogram by forming 2`*`2 scatter matrix, followed by 3`*`3 matrix of "sample size", "Median", and "Unemployment rate".

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```
